```python
columns = ['longitude']
for column in columns:
    sns.violinplot(x='latitude', y=column, data=employ)
    plt.xlabel('latitude')
    plt.ylabel(column)
    plt.title(f'{column} vs latitude')
    plt.show()
```

```python
columns = ['longitude']
for column in columns:
    sns.swarmplot(x='latitude', y=column, data=employ)
    plt.xlabel('latitude')
    plt.ylabel(column)
    plt.title(f'{column} vs latitude')
    plt.show()
```
